```python
# lendo o arquivo csv com pandas
df = pd.read_csv("us_macro_economic.csv")
```

```python
# exibindo as 5 primeiras linhas de dados
df.head()
```

```python
# renomeando as colunas e exibindo as primeiras 5 linhas de dados
df = df.rename(columns={"DATE":"Data","UNRATE(%)":"Desemprego(%)", "CONSUMER CONF INDEX":"Confiança do Consumidor", "PPI-CONST MAT.":"Compra Materiais de Construção", "CPIALLITEMS":"Preços ítens ao Consumidor", "INFLATION(%)":"Inflação(%)", "MORTGAGE INT. MONTHLY AVG(%)":"Média de Juros de Hipoteca(%)", "MED HOUSEHOLD INCOME":"Renda Familiar Média", "% SHARE OF WORKING POPULATION":"Parte da População Trabalhadora (18-60 anos)", "GDP PER CAPITA":"PIB Per Capita", "QUARTERLY REAL GDP":"PIB TRIMESTRAL REAL", 
                        "QUARTERLY GDP GROWTH RATE (%)	":"TAXA DE CRESCIMENTO TRIMESTRAL DO PIB (%)"})
df.head()
```

```python
# Converte "Data" para o formato "datetime"
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
# Pegando a inflação máxima de cada ano
df.groupby(df["Data"].dt.year)["Inflação(%)"].max()
```

```python
df["Data"].min()
```

```python
# Calculando a difença de dias
df["Diferença_dias"] = df["Data"] - df["Data"].min()
```

```python
# seleciona 5 dados aleatórios da tabela
df.sample(5)
```

```python
# inflação março de 2019
inflacao_marco_2019 = df.loc[(df["Data"].dt.year == 2005) & (df["Data"].dt.month == 5)]
```

```python
inflacao_marco_2019
```

```python
# Criando uma nova coluna para ano
df["Ano"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
# Inflação máxima de cada ano
df.groupby("Ano")["Inflação(%)"].max()
```

```python
# Renda familiar média
df["Renda Familiar Média"].mean()
```

```python
# 5 maiores taxas de inflação por ano
df.nlargest(5, "Inflação(%)")
```

```python
# 5 menores taxas de inflação por ano
df.nsmallest(5, "Inflação(%)")
```

```python
# soma dos PIBs por ano
df.groupby("Ano")["PIB Per Capita"].sum()
```

```python
# Visualização dos Dados
df["Inflação(%)"].value_counts(ascending=False)
```

```python
# Gráfico de Inflação máxima por ano
df.groupby("Ano")["Inflação(%)"].max().plot.bar();
```
